```python
# Importações de Bibliotecas Necessárias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from datetime import datetime
from datetime import timedelta  
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import scale
```

```python
# Load Generated Dataset (From DataGenerator_v5)
print("Reading CSV...")
df = pd.read_csv('generated_dataset_article_grouped_hour.csv', dayfirst=True, parse_dates=['timeFrom'], index_col='timeFrom')
print("Done!")
df
```

```python
# Split Important Time Variables 
df['hour'] = df.index.hour
#df['month'] = df.index.month
#df['weekday'] = df.index.weekday
#df['isRain'] = df['isRain'].astype('bool')
#df['isHoliday'] = df['isHoliday'].astype('bool')
#df['isWeekday'] = df['isWeekday'].astype('bool')
df['regpark'] = df['parking'] + ' - ' + df['region']

# Remove Unnecessary variables
df = df.drop(columns=['region','parking'])

# Show DataFrame
df
```

```python
# Filter just one Parking and Remove String Column
selected_parking = 'Bus Station - Main'

#df.reset_index()
#df = df.drop(columns=['timeFrom'])

df = df[df['regpark'] == selected_parking]
df = df.drop(columns=['regpark'])
df
```

```python
import pandas as pd
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=192)
y = kmeans.fit_predict(df[['isWeekday', 'isRain', 'isHoliday', 'hour']])
df['Cluster'] = y

print(df.head())
```

```python
train_size = 90
test = int(len(df) * ((100 - train_size) / 100))

df_train = df.iloc[:-test]
df_test = df.iloc[-test:]

print("Train: ",len(df_train))
print("Test: ",len(df_test))
```

```python
def predict_by_average(train_df, predict_df):
    predict_parkings = []
    for row, index in predict_df.iterrows():
        df_mean = train_df.copy()
        df_mean = df_mean[df_mean['Cluster'] == index['Cluster']]
        predict_parkings.append(df_mean['TotalParkings'].mean())

    predict_df['TotalParkingsPredict'] = predict_parkings
    return predict_df
```

```python
print("Running...")
predict_df = predict_by_average(df_train, df_test)
print("Done!")
```

```python
predict_df
```

```python
df3_dif = {'difference': [], 'proximity': []}

for row, index in predict_df.iterrows():
    prox_math = 100 - abs(round((((index['TotalParkingsPredict'] * 100) / index['TotalParkings'])) - 100, 2))
    df3_dif['proximity'].append(prox_math)
    
    dif_math = round((((index['TotalParkingsPredict'] * 100) / index['TotalParkings'])) - 100, 2)
    df3_dif['difference'].append(dif_math)
        
predict_df['PredictProximity'] = df3_dif['proximity']
predict_df['PredictDifference'] = df3_dif['difference']

predict_df
```
